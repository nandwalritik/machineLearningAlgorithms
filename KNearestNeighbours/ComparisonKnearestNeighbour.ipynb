# Own Implementation

```python
import KnearestNeighbours as KNN 
import numpy  as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.utils import shuffle
from sklearn.datasets import load_iris
```

```python
def normalize(X):
	return (X-np.mean(X))/(np.std(X))
```

```python
iris 		= 	load_iris()

X           =   iris.data
y 			=	iris.target

X,y = shuffle(X,y,random_state=0)	

sz   		=   iris.data.shape[0]
tsz			=   (int)(0.3*sz)
trainsz		=   sz-tsz
X_train,X_test,y_train,y_test = X[:trainsz,:],X[trainsz:,:],y[:trainsz],y[trainsz:]
```

```python
train = KNN.KNearestNeighbours()
train.setter(X_train,y_train,5)

test  = KNN.KNearestNeighbours()
test.setter(X_test,y_test)
```

```python
print("Accuracy as per my own class based implementation :- ",str(train.accuracy(test)))
```

```python
plt.scatter(X_train[:,2]*X_train[:,3],X_train[:,0]*X_train[:,1],c=y_train[:])
plt.xlabel("Product of petal length to petal width")
plt.ylabel("Product of sepal length to sepal width")
plt.title("Train Set")
plt.show()
```

```python
plt.scatter(X_test[:,2]*X_test[:,3],X_test[:,0]*X_test[:,1],c=train.predict(X_test))
plt.title("Test set(class based Implementation)")
plt.xlabel("Product of petal length to petal width")
plt.ylabel("Product of sepal length to sepal width")
plt.show()
```

# Scikit-Learn Implementation


```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
print("Confusion matrix :- \n",cm)
print("Accuracy of Scikit-Learn implementation :- ",100*accuracy_score(y_test,y_pred))
```

```python
plt.scatter(X_test[:,2]*X_test[:,3],X_test[:,0]*X_test[:,1],c=y_pred[:])
plt.title('Test Set as per Scikit-Learn')
plt.xlabel("Product of petal length to petal width")
plt.ylabel("Product of sepal length to sepal width")
plt.show()
```

# Own Implementation

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import KMeansClustering as KMC
```

```python
data  =  pd.read_csv('clustering.csv')
X = data.loc[:,["LoanAmount","ApplicantIncome"]].values
```

```python
train = KMC.KMeansClustering()
train.setter(X)
train.clustering(3)
```

```python
print("k cluster centroids as per my implementaion \n"+str(train.k_clusters_()))
```

```python
prediction = train.predict(X)
plt.scatter(X[:,0],X[:,1],c=prediction)
plt.title("Clusters as per my Implementation")
plt.show()
```

# Scikit-Learn Implementation

```python
from sklearn.cluster import KMeans 
model = KMeans(n_clusters=3,random_state = 0)
model.fit(X)
```

```python
print("k cluster centroids as per Scikit-Learn\n"+str(model.cluster_centers_))
```

```python
predicted = model.predict(X)
plt.scatter(X[:,0],X[:,1],c=predicted)
plt.show()
```
